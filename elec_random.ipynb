```python
#電気＋気温テスト
# ランダムフォレスト 季節モデル
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
#from sklearn.metrics import accuracy_score
import warnings
from sklearn.utils.testing import all_estimators
from sklearn.ensemble import RandomForestRegressor
#from sklearn.linear_model import LogisticRegression
from sklearn import preprocessing
import warnings
import time
warnings.filterwarnings('ignore')
#パラメータ設定
kaisuu = 1000
fukasa = 10
x_list=["Firepower","HydraulicPower","Biomass","SolarPower","WindPower","Pomping","Interconnection","Average","Temperature","Soloar","Wind","Humidity","High_temp","Low_temp"]
readCsv="Spring.csv"
line=237
testSize=60
repeat=int((line-testSize)/20)+1

#データフレームを一つにまとめる変数の初期化
result=pd.DataFrame()
for i in range(1,repeat):
    tmp=pd.DataFrame()
    # 景気データの読み込み
    #elec_data = pd.read_csv("elec_predict_for_random.csv",encoding="shift_jis")
    #elec_data = pd.read_csv("elec_predict_for_random.csv")
    elec_data = pd.read_csv(readCsv)
    #elec_data = pd.read_csv("kakunin.csv")
    elec_data=elec_data.drop(range(0,line-i*20-testSize))
    # データをラベルと入力データに分離する 
    y = elec_data.loc[:,"High"]
    x = elec_data.loc[:,["Firepower","HydraulicPower","Biomass","Solar power","Wind power","Pomping","Interconnection","Average","Temperature","Soloar","Wind","Humidity","High_temp","Low_temp"]]
    # 学習用とテスト用に分離する
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = testSize, train_size = 20*i, shuffle = False)
    #木を作って学習させて評価
    clf = RandomForestRegressor(max_depth=fukasa,n_estimators=kaisuu,random_state=False)
    #木を学習
    clf.fit(x_train, y_train)
    #予測値算出
    y_train_pred = clf.predict(x_test)
    t=pd.DataFrame(y_train_pred)
    #axis =0 縦に結合、axis=1は横に結合
    tmp=pd.concat([tmp,t],axis=0)
    result=pd.concat([result,tmp],axis=1)
    #寄与度
    print("モデル化期間"+str(20*i)+"日DONE\n")
fti = clf.feature_importances_    
import datetime
d_today = datetime.date.today()
year=d_today.year
month=d_today.month
day=d_today.day
today_now=str(year) + str(month) + str(day)
mode="Spring_入力"+str(len(x.columns))+"_month_"
name= today_now+mode+"学習回数"+str(kaisuu)+"深さ"+str(fukasa)
result.to_csv(name+'_elec_result.csv',index=False,header=False)
#寄与度
for j, feat in enumerate(x_list):
    print('\t{0:20s} : {1:>.2f}%'.format(feat, fti[j]*100))
```
