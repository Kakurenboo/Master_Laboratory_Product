```python
#sort
import pandas as pd
import warnings
import numpy as np
warnings.filterwarnings('ignore')
sort_data = pd.read_csv("sort_before_winter.csv",header=None)
#print(type(sort_data))

#Dataframe_to_リスト変換
sort_ls = sort_data.values.tolist()
sort_num = sort_data.values.tolist()
#設定箇所コ↑コ↓
#print(sort_ls)
for i in range(len(sort_ls)):
    for j in range(len(sort_ls[0])):
        sort_num[i][j]=j+1
for k in range(0,len(sort_ls)):
    for i in range(0,len(sort_ls[k])-1):
        for j in range(len(sort_ls[k])-1,i,-1):
            if sort_ls[k][j-1] <sort_ls[k][j]:
                tmp=sort_ls[k][j]
                sort_ls[k][j]=sort_ls[k][j-1]
                sort_ls[k][j-1]=tmp

                p=sort_num[k][j]
                sort_num[k][j]=sort_num[k][j-1]
                sort_num[k][j-1]=p
sort_ls.extend(sort_num)
df = pd.DataFrame(sort_ls)
print("---sort後---")
df.to_csv('sort_after_winter.csv',index=False,header=False)
```
